## 2. RAG + Reranker Architecture

Add a **reranker** between retrieval and final context selection.

**Key idea:**

- Initial retriever returns, say, 50 candidates
- Reranker (cross-encoder / LLM-based) scores them
- Top-k reranked chunks go to the LLM

**Diagram (text):**

```text
[User] 
  -> [API]
  -> [Retriever] ---> [Vector DB]
                      (top 50)
  -> [Reranker]  ---> (re-rank & filter top 5â€“10)
  -> [LLM Answer]
  -> [User + sources]
```

This design:

- Improves **precision** of final context
- Enables **hybrid** setups (dense + sparse)



## 3. Hybrid RAG (Dense + Sparse)

Combine:

- Dense embeddings (semantic search)
- Sparse search (BM25 / keyword / full-text)

Textual architecture:

```text
             +--> [Dense Retriever] ----+
[User Query]                             +--> [Fusion / Merger] -> [Top-k] -> [LLM]
             +--> [Sparse Retriever] ---+
```

Fusion can be:

- simple (interleave, average scores)
- learned (model to weight sources)



## 4. Agentic RAG Architecture

Connect to your **Agents Universe**.

**Idea:** An agent controls RAG tools:

- decides when to call retriever
- refines queries
- may do multi-hop retrieval

Textual architecture:

```text
[User]
  |
  v
[Agent (ReAct / Planner)] 
  |
  +--> (TOOL) search_knowledge_base() ---> [Vector / RAG backend]
  |                                          |
  |                                          v
  |                                    [Retrieved Chunks]
  |
  +--> (THOUGHT) "Is this enough?"
  |
  +--> (TOOL) search_knowledge_base() again, if needed
  |
  v
[LLM / Agent Final Answer] -> [User + citations]
```



## 5. RAG + MCP Architecture

Integrate **MCP (Model Context Protocol)** from your MCP Universe.

**Key shift:**

- RAG backends are exposed as MCP tools
- Agents call RAG via `mcp_call` instead of direct DB calls

Textual diagram:

```text
[User]
  |
  v
[Agent / Orchestrator]
  |
  +--> (mcp_call: server="rag_server", tool="search")
  |         |
  |         v
  |    [MCP RAG Server] ---> [Vector DB / Indexes]
  |
  v
[LLM Answer + Sources]
```

Benefits:

- Standardized interface for multiple RAG backends
- Same tools can be used by:
  - agents
  - copilots
  - CLI tools
