## 1. Why Domain-Specific RAG Patterns?

Different domains have:

* Different **document structures** (PDFs vs. structured data)
* Different **risk levels** (critical vs. low-stakes)
* Different **regulators** and **compliance rules**
* Different **evaluation needs** (accuracy thresholds, escalation rules)

RAG for:

* **Finance**: often regulated, numbers & rules heavy
* **Health**: safety-critical, must not replace clinicians
* **Legal**: nuanced, jurisdiction-specific, requires expert oversight

This notebook gives **patterns** you can adapt in your projects.

---

## 2. Shared Building Blocks Across Domains

All three domains commonly share:

1. **Document ingestion**

   * PDFs, web pages, internal docs, APIs
   * Versioning and provenance

2. **Chunking & embeddings**

   * Domain-aware chunking (by section, clause, guideline)
   * Embeddings tuned for text vs. code vs. tables

3. **Vector store & metadata**

   * Tag by domain, date, jurisdiction, product, etc.

4. **Retrieval**

   * Basic (RAG v1) vs. hybrid + reranker (RAG v2)
   * Sometimes graph-based (GraphRAG) for complex dependencies

5. **RAG Engine / Orchestrator**

   * Simple pipeline, or Agentic RAG with planner + tools

6. **Guardrails & Safety**

   * Filters, disclaimers, escalation to human experts
